```python


#### CENTER OF MASS ####
com = dataset[1][0].xs('Center of mass', level='bodyparts', axis=1)
# Discard Likelihood
com = com.drop('likelihood', level=1, axis=1)

# Compute distnaces between com
com_x = com.xs('x', level=1, axis=1).to_numpy()
com_y = com.xs('y', level=1, axis=1).to_numpy()

dist_x = com_x[:,0] - com_x[:,1]
dist_y = com_y[:,0] - com_y[:,1]
dist_com = np.sqrt(dist_x**2 + dist_y**2)
# Drop Nans
mask = np.isnan(dist_com)
dist_com = dist_com[~mask]


print(" Output has Nans: ", np.isnan(dist_com).any())

#dataset[1][1].iloc[:,1].to_numpy()[~mask]
dataset[1][1].shape[0]
```

```python

y = []
distance_center_of_mass = []
coords_dmd_null = []
outcome = []
for i in tqdm.tqdm(range(len(dataset)-1)):

    ### Prepare a dataset wich is simply all the points concatenated
    n_frames, features_coords = dataset[i][0].values.shape
    n_bp = features_coords // 3
    coords = dataset[i][0].values.reshape(n_frames*n_bp, 3)

    has_nan = np.isnan(coords).any(axis=1)
    coords[has_nan, :] = [0, 0, 0]
    coords = coords.reshape(n_frames, features_coords)

    if coords.shape[0] != dataset[i][1].shape[0]:
        print("Warrning: ", i, " ", coords.shape[0], " ", dataset[i][1].shape[0])
    coords_dmd_null.append(coords)
    outcome.append(dataset[i][1].iloc[:,1].to_numpy())

    #### CENTER OF MASS ####
    com = dataset[i][0].xs('Center of mass', level='bodyparts', axis=1)
    # Discard Likelihood
    com = com.drop('likelihood', level=1, axis=1)

    # Compute distnaces between com
    com_x = com.xs('x', level=1, axis=1).to_numpy()
    com_y = com.xs('y', level=1, axis=1).to_numpy()

    dist_x = com_x[:,0] - com_x[:,1]
    dist_y = com_y[:,0] - com_y[:,1]
    dist_com = np.sqrt(dist_x**2 + dist_y**2)
    mask = np.isnan(dist_com)
    distance_center_of_mass.append(dist_com[~mask])

    y.append(dataset[i][1].iloc[:,1].to_numpy()[~mask])

  

    if False:
        #### NOSE-NOSE ####
        nose = dataset[i][0].xs('Nose', level='bodyparts', axis=1)
        # Discard Likelihood
        nose = nose.drop('likelihood', level=1, axis=1)

        # Compute distnaces between Nnose
        nose_x = nose.xs('x', level=1, axis=1)
        nose_y = nose.xs('y', level=1, axis=1).t

        dist_x = nose_x[:,0] - nose_x[:,1]
        dist_y = nose_y[:,0] - nose_y[:,1]
        dist_nose = np.sqrt(dist_x**2 + dist_y**2)
        # Drop Nans
        dist_nose = dist_nose[~np.isnan(dist_nose)]

        #### TAIL-TAIL ####
        tail = dataset[i][0].xs('Tail_base', level='bodyparts', axis=1)
        # Discard Likelihood
        tail = tail.drop('likelihood', level=1, axis=1)

        # Compute distnaces between Tail
        tail_x = tail.xs('x', level=1, axis=1).to_numpy()
        tail_y = tail.xs('y', level=1, axis=1).to_numpy()

        dist_x = tail_x[:,0] - tail_x[:,1]
        dist_y = tail_y[:,0] - tail_y[:,1]
        dist_tail = np.sqrt(dist_x**2 + dist_y**2)
        # Drop Nans
        dist_tail = dist_tail[~np.isnan(dist_tail)]

        #### TAIL-NOSE ####
        dist_x_0 = nose_x[:,0] - tail_x[:,1]
        dist_y_0 = nose_y[:,0] - tail_y[:,1]
        dist_tail_nose_0 = np.sqrt(dist_x_0**2 + dist_y_0**2)
        # Drop Nans
        dist_tail_nose_0 = dist_tail_nose_0[~np.isnan(dist_tail_nose_0)]

        dist_x_1 = nose_x[:,1] - tail_x[:,0]
        dist_y_1 = nose_y[:,1] - tail_y[:,0]
        dist_tail_nose_1 = np.sqrt(dist_x_1**2 + dist_y_1**2)
        # Drop Nans
        dist_tail_nose_1 = dist_tail_nose_1[~np.isnan(dist_tail_nose_1)]
```

## Sniff Total resident active

We need first understand the differences between both mice. Not an eassy task!

I'm going to take as features the nose coordinate and all the border coordinates of the other mouse. I'm going to use the distance between the nose and the center of mass, and the distance between the nose and the tail.

Let's create the features that we want. By convention let's choose the first mouse as the resident.

```python
# For each video, create the features.

X_train = pd.DataFrame()

y_list = []

X_train = pd.DataFrame()
frames_per_train = []

for test in dataset:
    nose_coords_R = test[0].xs('Nose', level='bodyparts', axis=1).xs('individual2', level='individuals', axis=1)

    individual_2_coords = test[0].xs('individual1', level='individuals', axis=1)

    # Substract the nose coordinates of the first individual to all the bodyparts of the second one exept from likelihood 
    dist_coords = individual_2_coords - nose_coords_R
    
    # Drop Likelihood
    dist_coords = dist_coords.drop('likelihood', level=1, axis=1)



    # rename the columns, add 'dist_' to the name of the bodypart
    dist_coords.columns = pd.MultiIndex.from_tuples([(f'dist_{col[0]}', col[1]) for col in dist_coords.columns])

    # Max distance in absolute value of all the bodyparts that is not nan
    max_dist = dist_coords.abs().max().max()

    # rename x coordinate as dist
    #dist_coords = dist_coords.rename(columns={'x': 'dist'})

    dist_coords = dist_coords**2

    # Compress x and y coordinates into a single column by doing x + y
    for bodypar in dist_coords.columns.levels[0]:
        dist_coords[bodypar] = dist_coords[bodypar].sum(axis=1)

    # Drop the y column
    dist_coords = dist_coords.drop('y', level=1, axis=1)

    # forget about the level of the columns
    dist_coords.columns = dist_coords.columns.droplevel(1)
    # sqrt of the sum of the squares
    dist_coords = np.sqrt(dist_coords)

    frames_per_train.append(dist_coords.shape[0])
    y_list.append(test[1]['sniff total resident active'].values)

    # Add to a global dataframe that will contain all the data
    X_train = pd.concat([X_train, dist_coords], axis=0, ignore_index=True)

y_train = np.concatenate(y_list)
```
