```python
import os
import numpy as np
import pandas as pd
```

```python
folder_gt_path = r'c:\Users\jalvarez\Documents\Data\DataSet_MDX52_INES\GT'
folder_pred_path = r'c:\Users\jalvarez\Documents\Data\DataSet_MDX52_INES\outputs'
```

```python
# Load the GT
files_gt = os.listdir(folder_gt_path)
files_gt = sorted(files_gt)
files_gt
```

```python
files_pred = os.listdir(folder_pred_path)
files_pred = sorted(files_pred)
files_pred
```

```python
gt_df = pd.DataFrame()
pred_df = pd.DataFrame()

for file in files_gt:
    print(file)
    df_gt = pd.read_csv(os.path.join(folder_gt_path, file))
    # Discard the first 2 frames and last 2 frames
    df_gt = df_gt.iloc[2:-2, :]
    gt_df = pd.concat([gt_df, df_gt])

    for file_pred in files_pred:

        if file.split('.')[0] == file_pred.split('_')[0] + '_' + file_pred.split('_')[1]:
            print(file_pred.split('_')[0] + '_' + file_pred.split('_')[1])
            df_pred = pd.read_csv(os.path.join(folder_pred_path, file_pred))
            # Discard the first 2 frames and last 2 frames
            # check if the length is the same
            if len(df_gt) != len(df_pred):
                print(f'Length of {file} is different from {file_pred}')
                break
            pred_df = pd.concat([pred_df, df_pred])
            
```

```python
# Restore index
gt_df.reset_index(drop=True, inplace=True)
pred_df.reset_index(drop=True, inplace=True)
```

```python
# Reorder the columns
gt_df = gt_df.reindex(sorted(gt_df.columns), axis=1)
pred_df = pred_df.reindex(sorted(pred_df.columns), axis=1)
```

```python
gt_df
```

```python
pred_df
```

```python


# Calculate the accuracy
# Convert to numpy
gt = gt_df.to_numpy()
pred = pred_df.to_numpy()

print(gt.shape)
print(pred.shape)
```

```python
# Calculate the accuracy for each class
acc = np.mean(gt == pred, axis=0)
print(' Accuracy per class:', acc)

# Calculate the global accuracy
global_acc = np.mean(gt == pred)
print('Global Accuracy:', global_acc)
```

```python
# import the confusion matrix

from sklearn.metrics import confusion_matrix
```

```python
# Calculate the confusion matrix for each class
conf_matrix = np.zeros((gt.shape[1], 2, 2))
for i in range(gt.shape[1]):
    conf_matrix[i] = confusion_matrix(gt[:, i], pred[:, i])

conf_matrix
```

```python
# calculate the percentage CM 
for i, cm in enumerate(conf_matrix):
    conf_matrix[i] = cm/cm.sum(axis=1)[:, np.newaxis]
    
```

```python
conf_matrix
```

```python
# Plot the confusion matrix for each class
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style and context for the plot
sns.set_theme(style="whitegrid", context="talk")

# Subplot for 7 classes
fig, axs = plt.subplots(4, 2, figsize=(16, 24))
fig.suptitle('Confusion Matrix per Class', fontsize=24, fontweight='bold')

for i in range(4):
    for j in range(2):
        if i*2 + j >= gt_df.shape[1]:
            break
        sns.heatmap(conf_matrix[i*2 + j], annot=True, fmt='.2f', cmap='Blues', ax=axs[i, j], cbar=False)
        axs[i, j].set_title(f'{gt_df.columns[i*2 + j]}', fontsize=16, fontweight='bold')
        axs[i, j].set_xlabel('Predicted', fontsize=14)
        axs[i, j].set_ylabel('True', fontsize=14)

plt.show()
```
